```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=keras.datasets.fashion_mnist
(train_x,train_y),(test_x,test_y)=data.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
train_x.shape
train_x=train_x/255.0
test_x=test_x/255.0
train_x.shape
```

```python
plt.imshow(train_x[0],cmap=plt.cm.binary)
plt.show
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense,Dropout, Conv2D,MaxPooling2D,Activation,LeakyReLU
from tensorflow.keras.callbacks import TensorBoard
```

```python
tensorboard=TensorBoard(log_dir='logs/mnist_dataset')
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(256))
model.add(LeakyReLU(alpha=0.05))
model.add(Dense(128))
model.add(LeakyReLU(alpha=0.05))
model.add(Dense(10))
model.add(Activation('sigmoid'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

history=model.fit(train_x,train_y,batch_size=128,epochs=10,callbacks=[tensorboard],validation_data=(test_x,test_y))
```

```python
test_loss,test_acc=model.evaluate(test_x,test_y,verbose=1)
print("test accuracy=",test_acc)
```

```python
predictions=model.predict(test_x)
[print(class_names[np.argmax(i)]) for i in predictions]
```

```python
model.save('mnist_fashion.model')
```
